```python
item_price={"Milk":17,"Meat":180,"Pen":12,"Headphone":80,"Chips":8,"Rice":20,"Tea":10}

cart=[]


while True:
  cmd=input("Enter a comand:   1-Add item   2-Remove item   3-Print the total amount   4-Clear the cart:  ").lower()
  if cmd == "add item" or cmd == "1":
    item=input('''  "Milk":17   "Meat":180     "Pen":12    "Headphone":80    "Chips":8    "Rice":20    "Tea":10  ''' ).lower()
    cart.append(item)
    print("items in cart",cart)
  elif cmd == "remove item" or cmd == "2":
    print("items in cart",cart)
    item=input("Which one do you want to remove?").lower()
    cart.remove(item)
    print("items in cart",cart)
  elif cmd == "print the total amount" or cmd == "3":
    total_amount=0
    for item in cart:
      total_amount+=item_price[item.capitalize()]
    print("Total amount:", total_amount)
  elif cmd == "clear the cart" or cmd == "4":
    cart = []
    print("Cart cleared.")
  else:
    print("Invalid command.")
```
