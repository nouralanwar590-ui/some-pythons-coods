```python
N1=int(input("Enter the first num: "))
N2=int(input("Enter the second num: "))

import math_assistant

while True:
  optoin=input("Enter the option (+, -, *, /): ").lower()
  if optoin=='+' or optoin=="add":
   math_assistant.add(N1,N2)
  elif optoin=='-' or optoin=="sup":
   math_assistant.sub(N1,N2)
  elif optoin=='*'or optoin=="multiplication":
   math_assistant.Multiplication(N1,N2)
  elif optoin=='/'or optoin=="division":
   math_assistant.Division(N1,N2)
  else:
    print("Invalid option")
```
