```python
import pandas as pd
```

```python
house={
    "Price":[2000,1500,2700,3000],
    "No.Rooms":[2,1,2,3],
    "Size":[145,130,175,190],
    "Floor Number":[2,3,4,1]
}
ratings=pd.DataFrame(house)
print(ratings)
```

```python
new_house=pd.Series([2500,3,170,2])
recommendation=ratings.corrwith(new_house)
print(recommendation)
```
