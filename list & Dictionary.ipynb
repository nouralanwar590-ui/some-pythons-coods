```python
item_price={"milk":30,"meat":200,"pen":12}

cart=[]


while True:
  cmd=input("Do you wantto add item?: ").lower()
  if cmd=="add item":
    item=input('''   "milk":30,   "meat":200 ,   "pen":12   ''' ).lower()
    cart.append(item)


  print("items in cart")
  for i in cart:
    print("item: {}  .    price: {} . ".format(i, item_price[i]))
```
