```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('fruit_data_with_colors.csv', sep=r"\s+")
print(df.columns)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
print(df.columns)
```

```python
x = df[['mass', 'width', 'height', 'color_score']]
y = df['fruit_label']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.7,random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
print("array",y_test.to_numpy())
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy:{accuracy: .3f}")
```

```python
accuracy = knn.score(x_test, y_test)
print("Model Accuracy:", accuracy)
```

```python
fruit=[[150, 7.5, 7.2, 0.65]]
fruit_labels = df[['fruit_label','fruit_name']].drop_duplicates().set_index('fruit_label')['fruit_name']
pred=knn.predict(fruit)
print(fruit_labels[pred[0]])
```

```python
import numpy as np

new_sample = np.array([[150, 7.5, 7.2, 0.65]])

prediction = knn.predict(new_sample)[0]

fruit_dict = df[['fruit_label','fruit_name']].drop_duplicates().set_index('fruit_label')['fruit_name']
print("Predicted Fruit:", fruit_dict[prediction])
```
