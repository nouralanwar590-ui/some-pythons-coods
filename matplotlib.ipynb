```python
import matplotlib.pyplot as plt

year=[2017,2018,2019,2020,2021,2022]
expense=[100000,120000,128000,110000,130000,150000]
profit=[10000,20000,18000,21000,20000,27000]

startup ={"year":year, "expense":expense, "profit":profit}

```

```python
import pandas as pd
df =pd.DataFrame(startup)
df.to_csv("startup.csv")
df =pd.read_csv("startup.csv")
df

```

```python
plt.plot(df["year"],df["expense"])
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(df["year"],df["profit"])
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.bar(df["year"],df["expense"])
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(df["year"],df["profit"]  ,label="profit")
plt.plot(df["year"],df["expense"] ,label="expense")
plt.legend(loc="upper left")
```
