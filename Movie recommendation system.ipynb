```python
import pandas as pd
```

```python
movies=pd.read_csv("/content/Netflix_Dataset_Movie.csv")
movies.head(10)
```

```python
ratings =pd.read_csv("/content/Netflix_Dataset_Rating.csv")
ratings.head(10)
```

```python
movies_ratings= pd.merge(movies,ratings)
movies_ratings.head(10)
```

```python
movies_ratings= movies_ratings.pivot_table(index=['User_ID'],columns=['Name'],values='Rating')
movies_ratings.head()
```

```python
movies["Name"][movies["Name"].str.contains("Wild")]
```

```python
wild_ratings=movies_ratings["Wild Things"]
wild_ratings.head()
```

```python
similar_movies=movies_ratings.corrwith(wild_ratings)
similar_movies.head()
```

```python
similar_movies.sort_values(ascending=False)
```
