```python
password = "0000"
balance = 3000
is_running = True

while is_running:
    Password1 = input("Enter your password: ")

    if password == Password1:
        print("Password is correct\n")

        while True:
            print("--- ATM Menu ---")
            option = input("Choose an operation: 1-Balance | 2-Make deposit | 3-Make withdraw: ")


            if option == "1":
                print(f"Your balance is: ${balance}")


            elif option == "2":
                deposit = int(input("How much do you want to deposit? "))
                balance += deposit
                print(f"Your new balance is: ${balance}")


            elif option == "3":
                withdraw = int(input("How much do you want to withdraw? "))
                if withdraw > balance:
                    print("Operation failed: You don't have enough money.")
                else:
                    balance -= withdraw
                    print(f"Your new balance is: ${balance}")
            else:
                print("Invalid option!")


            print("-" * 20)
            again = input("Do you want another operation? (yes/no):").strip().lower()

            if again == "2" or again == "no":
                print("Thank you for using our ATM. Goodbye!")
                is_running = False
                break
            elif again == "1" or again == "yes":
                pass

    else:
        print("Password is incorrect. Please try again.\n")
```
