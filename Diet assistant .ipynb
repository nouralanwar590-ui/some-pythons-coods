```python
import pandas as pd

# Sample data
data = {
    'Name': ['Chicken Breast', 'Tofu', 'Salmon', 'Almonds', 'Broccoli'],
    'Calories': [165, 76, 208, 575, 55],
    'Protein': [31, 8, 20, 21, 3.7]
}

# Create the DataFrame
df = pd.DataFrame(data)

df
```

```python
df['Protein_per_Calorie'] = df['Protein'] / df['Calories']
df.sort_values(by='Protein_per_Calorie', ascending=False)
```

```python
high_protein_low_cal = df[(df['Protein'] > 15) & (df['Calories'] < 250)]
df
```
