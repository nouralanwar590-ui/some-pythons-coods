**Recommendatoin movie system **

```python
import pandas as pd
Movies={
"Spiderman":[3.5,5,1,4],
"James bond":[2.0,3.5,4,5],
"Titanic":[5.0,3.5,1.0,2.0]
}
ratings=pd.DataFrame(Movies)
print(ratings)
```

```python

new_ratings=pd.Series([1.0,5.0,2.5,4.0])
recommendation=ratings.corrwith(new_ratings)
print(recommendation)
```

```python
recommendation = recommendation.sort_values(by='Correlation', ascending=False)
print(recommendation)
```

```python
recommendation = pd.DataFrame(recommendation,columns=['Correlation']).reset_index()
print(recommendation)
```
